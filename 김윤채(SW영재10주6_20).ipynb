```python
#다
import matplotlib.pyplot as plt
import numpy as np
import csv
from enum import Enum
f = open('age.csv','r', encoding='utf8')
data = csv.reader(f)
next(data)
h_a = []
f_a = []
s1 = input('')
total = []
for row in data:
  if s1 in row[0]
    total.append(int(row[1]))
    for i in row[229:249]:
      f_a.append(int(i))
    for a in row[68:104]:
      h_a.append(int(a))
    break
def warn_number():
    return sum(f_a) / sum(h_a)
def chose_lowest_5():
  numbers = warn_number()
  lowest_five =  sorted(numbers)[:5]
  return lowest_five
print(chose_lowest_5())
print(str(s1) +' 전체 인구 : '+str(total[0]) + ' 20~39세 여성인구 : '+str(sum(f_a)) + ' 65세 이상 인구 : '+str(sum(h_a)) + ' 소멸위험지수 : '+str(warn_number()))

```

```python
import numpy as np
from sklearn import linear_model  # scikit-learn 모듈을 가져온다
regr = linear_model.LinearRegression()
X = [[164], [179], [162], [170]]  # 다중회귀에도 사용하도록 함
y = [53, 63, 55, 59]              # y = f(X)의 결과
regr.fit(X, y)
coef = regr.coef_           # 직선의 기울기
intercept = regr.intercept_ # 직선의 절편
score = regr.score(X, y)    # 학습된 직선이 데이터를 얼마나 잘 따르나
print("y =", coef, "* X + ", intercept)
print("The score of this line for the data: ", score)
```

```python
input_data = [ [180], [185] ]
result = regr.predict(input_data)
print(result)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
import csv
import matplotlib.pyplot as plt
f = open('출생아데이터.csv', 'r', encoding='utf8')
data = csv.reader(f)
next(data)
X = []
y = []
for row in data:
    y.append(int(row[1]))
    X.append(int(row[0]))
X = np.array(X).reshape(-1, 1)
y = np.array(y)
train_input, test_input, train_target, test_target = train_test_split(X, y, random_state=42)
knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(train_input, train_target)
distances, indexes = knr.kneighbors([[100]])
plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
plt.xlabel('year')
plt.ylabel('인구수')
plt.show()

```
